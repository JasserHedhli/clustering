```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("/content/Mall_Customers.csv")
```

```python
df.head()
```

```python
# @title Gender vs Spending Score (1-100)

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(df['Gender'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(df, x='Spending Score (1-100)', y='Gender', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
df.info()
```

```python
df.columns
```

```python
dfa = df[['Age', 'Annual Income (k$)']]
dfa
```

```python
scdfa = StandardScaler()
dfa_std = scdfa.fit_transform(dfa.astype(float))
```

```python
dfa_std
```

```python
cluster_range = [2,3,4,5,6,7,8,9,10]
inertie = []

for c in cluster_range:
  kmeans = KMeans(init="k-means++", n_clusters = c).fit(dfa_std)
  inertie.append(kmeans.inertia_)

plt.figure()
plt.plot(cluster_range, inertie, marker='o')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3).fit(dfa_std)
labels = kmeans.labels_
new_dfa = pd.DataFrame(data = dfa_std, columns = ['age', 'revenue_annuel'])
new_dfa['labels_kmeans'] = labels
new_dfa
```

```python
new_dfa[new_dfa['labels_kmeans']==1]
```

```python
fig,ax=plt.subplots(figsize=[10,7])
plt.scatter(new_dfa['revenue_annuel'][new_dfa['labels_kmeans']==0],new_dfa['age'][new_dfa['labels_kmeans']==0],
            color='blue',s=100,linestyle='--')
plt.scatter(new_dfa['revenue_annuel'][new_dfa['labels_kmeans']==1],new_dfa['age'][new_dfa['labels_kmeans']==1],
            color='red',s=100,linestyle='--')
plt.scatter(new_dfa['revenue_annuel'][new_dfa['labels_kmeans']==2],new_dfa['age'][new_dfa['labels_kmeans']==2],
            color='green',s=100,linestyle='--')

plt.show()
```
